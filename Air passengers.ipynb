```python
Airports=['PARIS-CHARLES DE GAULLE airport','PARIS-ORLY airport','NICE-COTE D AZUR airport',
          'ADOLFO SUAREZ MADRID-BARAJAS airport','BARCELONA/EL PRAT airport','PALMA DE MALLORCA airport','FRANKFURT/MAIN airport',
          'MUENCHEN airport','BERLIN-BRANDENBURG airport','ROMA/FIUMICINO airport','MILANO/MALPENSA airport','BERGAMO/ORIO AL SERIO airport'
          ,'DUBLIN airport','CORK airport','SHANNON airport']
df_passenger_airports=df_passenger_airport[df_passenger_airport['Airport'].isin(Airports)] 
df_national_passenger_airports=df_national_passenger[df_national_passenger['Airport'].isin(Airports)] 
```

```python
# Función para asignar el país basado en el aeropuerto
def assign_country(row):
    airport_mapping = {
        'Spain': ['ADOLFO SUAREZ MADRID-BARAJAS airport', 'BARCELONA/EL PRAT airport', 'PALMA DE MALLORCA airport'],
        'France': ['PARIS-CHARLES DE GAULLE airport', 'PARIS-ORLY airport', 'NICE-COTE D AZUR airport'],
        'Germany': ['FRANKFURT/MAIN airport', 'MUENCHEN airport', 'BERLIN-BRANDENBURG airport'],
        'Italy': ['ROMA/FIUMICINO airport', 'MILANO/MALPENSA airport', 'BERGAMO/ORIO AL SERIO airport'],
        'Ireland': ['DUBLIN airport', 'CORK airport', 'SHANNON airport']
    }

    for country, airports in airport_mapping.items():
        if row['Airport'] in airports:
            return country
    return np.nan

df_passenger_airports['Country'] = df_passenger_airports.apply(assign_country, axis=1)

print(df_passenger_airports[['Airport', 'Country']])
```

```python
df_passenger_summarized = df_passenger_countries.melt(id_vars=['Country'], var_name='Year', value_name='AirTraffic')
df_passenger_summarized['Year'] = df_passenger_summarized['Year'].str.split('-').str[0]
df_passenger_summarized['AirTraffic'] = pd.to_numeric(df_passenger_summarized['AirTraffic'])

app = dash.Dash(__name__)

years = df_passenger_summarized['Year'].unique()

app.layout = html.Div([
    html.H1("Tráfico Aéreo por País"),
    dcc.Graph(id='world-map'),
    dcc.Slider(
        id='year-slider',
        min=0,
        max=len(years) - 1,
        step=1,
        value=0,
        marks={i: year for i, year in enumerate(years)},
        tooltip={'placement': 'bottom', 'always_visible': True}
    ),
])

@app.callback(
    Output('world-map', 'figure'),
    [Input('year-slider', 'value')]
)
def update_map(selected_year):
    selected_year = years[selected_year]

    df_yearly_sum = df_passenger_summarized[df_passenger_summarized['Year'] == selected_year]
    df_yearly_sum = df_yearly_sum.groupby('Country')['AirTraffic'].sum().reset_index()

    fig = px.bar(
        df_yearly_sum,
        x='Country',
        y='AirTraffic',
        title=f'Tráfico Aéreo por País en {selected_year}',
        labels={'Country': 'País', 'AirTraffic': 'Tráfico Aéreo'},
        hover_data=['AirTraffic'],  # Muestra el valor al pasar el mouse
        text='AirTraffic',  # Muestra el valor en las barras
    )

    fig.update_layout(showlegend=False)
    total_value = df_yearly_sum['AirTraffic'].sum()
    fig.add_annotation(
        x=5,
        y=-0.15,
        showarrow=False,
        text=f'Total: {total_value}',
        font=dict(size=12, color="black")
    )

    return fig

if __name__ == '__main__':
    app.run_server(debug=True, port=8052)
```

```python
def read_credentials(file_path='REDDIT_ACCESS.txt'):
    credentials = {}
    with open(file_path, 'r') as file:
        for line in file:
            key, value = line.strip().split(': ')
            credentials[key] = value
    return credentials

def authenticate_reddit():
    try:
        credentials = read_credentials()
        reddit = praw.Reddit(
            client_id=credentials['client_id'],
            client_secret=credentials['client_secret'],
            user_agent=credentials['user_agent']
        )
        print("Login successful.")
        return reddit
    except Exception as e:
        print(f"Error logging in: {e}")
        return None

def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment_score = analysis.sentiment.polarity
    
    # Classify the sentiment
    if sentiment_score > 0:
        return "Positive"
    elif sentiment_score < 0:
        return "Negative"
    else:
        return "Neutral"

def calculate_global_sentiment(search_results):
    total_sentiment = 0
    total_comments = 0

    for submission in search_results:
        submission.comments.replace_more(limit=None)
        comments = submission.comments.list()
        
        for comment in comments:
            if hasattr(comment, 'body'):
                text = comment.body
                sentiment = analyze_sentiment(text)
                
                # Incrementa el contador según la clasificación del sentimiento
                if sentiment == "Positive":
                    total_sentiment += 1
                elif sentiment == "Negative":
                    total_sentiment -= 1
                # No se hace nada si el sentimiento es "Neutral"

                total_comments += 1

    if total_comments > 0:
        global_sentiment = total_sentiment / total_comments
        return global_sentiment
    else:
        return None

def main():
    # Authenticate on Reddit
    reddit_instance = authenticate_reddit()

    # Check if login was successful
    if reddit_instance:
        airlines = ["AirFrance", "Lufthansa", "AerLingus", "Alitalia", "Iberia"]
        
        for airline in airlines:
            print(f"\nAnalyzing sentiment for {airline}:\n")
            
            # Get comments related to the airline
            subreddit = reddit_instance.subreddit("travel")
            search_results = subreddit.search(airline, sort="new", limit=20)

            # Calculate global sentiment
            global_sentiment = calculate_global_sentiment(search_results)

            if global_sentiment is not None:
                # Imprimir el resultado junto con la clasificación del sentimiento
                sentiment_label = "Positive" if global_sentiment > 0 else "Negative" if global_sentiment < 0 else "Neutral"
                print(f"Global sentiment for {airline}: {global_sentiment} ({sentiment_label})")
            else:
                print(f"No comments found for {airline}.")

    else:
        print("Login unsuccessful. Check credentials.")

if __name__ == "__main__":
    main()
```
