```python
import pandas as pd

df_passenger_country='Air passenger transport by reporting country.tsv'
df_passenger_airport='Air passenger transport by main airports in each reporting country.tsv'
df_national_passenger='National air passenger transport by reporting country.tsv'

df_passenger_country=pd.read_csv(df_passenger_country, sep='\t')
df_passenger_airport=pd.read_csv(df_passenger_airport, sep='\t')
df_national_passenger=pd.read_csv(df_national_passenger, sep='\t')
```

```python
df_passenger_country_Q = df_passenger_country['unit,tra_meas,tra_cov,schedule,geo\\time'].str.split(',', expand=True)

columns_Q= [col for col in df_passenger_country.columns if ('Q' in col) and (int(col[:4]) >= 2012) and (int(col[:4]) <= 2022)]

# Seleccionar las columnas del DataFrame original correspondientes a los trimestres
df_passenger_country_Q2 = df_passenger_country[columns_Q]

# Obtener la última columna del DataFrame original
country_column = df_passenger_country['unit,tra_meas,tra_cov,schedule,geo\\time'].str.split(',').str[-1]

# Concatenar las columnas 'geo\time' y las columnas de trimestres en un nuevo DataFrame
df_passenger_country_Q = pd.concat([country_column, df_passenger_country_Q2], axis=1)

# Cambiar el nombre de la última columna a 'Country'
df_passenger_country_Q = df_passenger_country_Q.rename(columns={country_column.name: 'Country'})

# Mostrar las primeras filas del DataFrame filtrado
df_passenger_country_Q.head()
```

```python
df_passenger_country_Q.columns
```

```python
df_country = pd.read_excel('Dictionary.xlsx')

# Merge para combinar los DataFrames usando la columna 'Country' como clave
df_final = pd.merge(df_passenger_country_Q, df_country, left_on='Country', right_on='Country', how='left')

# Reemplazar la columna 'Country' con la columna 'Nombre' (si está disponible)
df_final['Country'] = df_final['Name'].fillna(df_final['Country'])

# Eliminar las columnas adicionales ('Abreviatura' y 'Nombre') si lo deseas
df_final = df_final.drop(['Dictionary', 'Name'], axis=1, errors='ignore')

# Mostrar las primeras filas del DataFrame final
df_final.head()
```

```python
# Suponiendo que df_final es tu DataFrame final después de la operación de combinación
unique_country = df_final['Country'].unique()

# Mostrar los valores únicos
print(unique_country)
```

```python
include_countries= ['Spain', 'France', 'Ireland', 'Norway', 'Italy']

# Filtrar las filas que contienen los países especificados
df_passenger_by_country = df_final[df_final['Country'].isin(include_countries)]

# Mostrar las primeras filas del DataFrame filtrado
df_passenger_by_country.head()
```
